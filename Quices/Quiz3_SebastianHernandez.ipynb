### 1. (2.0)
Design a function that maps a sample in the ${\rm I\!R}^2$ input space to a feature space induced by the kernel $k(x,y)=(2<x,y>+1)^{2}$. 

```python
import numpy as np

def phi(x):
    '''
    x: a sample in the input space (numpy array of shape (2,))
    returns a vector in the feature space corresponding to the image of x (numpy array of shape (6,))
    '''
    #Se calcula cada factor del polinomio resultante del producto punto y se coloca como una columna de la matriz, si 
    #se tienen mas 2 de tuplas de datos, se colocara como una fila de matriz resultante
    return np.array([2*x[0]**2, 2*x[1]**2, 2*np.sqrt(2)*x[0]*x[1], 2*x[0], 2*x[1], 1],dtype='object')
```
