## 1.  (1.0)

Write a function that receives $P(X|Y)$ and $P(Y)$ and returns $P(X)$


````python
import numpy as np

def p1(P_XY, P_Y):
    '''
    P_XY: a numpy array of shape (m,n) where P_XY[i, j] = P(X=i | Y=j) 
    P_Y:  a numpy array of shape (n,) where P_Y[j] = P(Y=j) 
    
    Returns:
    A numpy array P_X[i] = P(X=i) 
    '''
    ### your code here
    #Primero se halla la probabilidad conjunta, es decir P(X,Y)
    PXY=np.multiply(P_XY,P_Y)
    #Ya con la probabilidad conjunta, se puede puede hallar la probabilidad total sumando en las dimensiones de la matriz,
    #es decir, se esta usando el hecho que P(X,Y[0])+P(X,Y[1])+P(X,Y[2])+P(X,Y[3])=P(X,1)=P(X)
    PX = PXY.sum(axis=1)
    PY = PXY.sum(axis=0)
    return PX
````

````python
## TEST_CELL
p_xy = np.array([[0.1, 0.7, 0.4], [0.9, 0.3, 0.6]])
p_y  = np.array([0.6, 0.4, 0.1])
print(p1(p_xy, p_y))
````

## 2.  (2.0)
Assume that $X$ is a random variable that takes values in the set $\{0,1, \dots, n-1\}$. 
Write a function that receives $P(X)$ and calculates $E[X^2] - E[X]^2$


````python
def p2(px):
    '''
    P_X: a numpy array of shape (n,) where P_X[i] = P(X=i)  
    
    Returns:
        E[Xˆ2] - E[X]^2
    '''
    ### your code here
    #Primero, se arma el vector referente a x
    x=np.array(range(0,(px.shape[0]))).T
    #Y luego se usa el hecho de que el valor esperado de una distribución discreta es igual a la sumatoria de
    #cada elemento multiplicado por su probabilidad, es decir, E[x]=sum(x*px)
    Ex2 =sum(np.multiply((np.power(x,2)),px))
    E2x=np.power(sum(np.multiply(x,px)),2)
    res = Ex2-E2x
    return res
````

## 3. (2.0)
Suppose you have a 5 different binary classifiers and the probability that a particular classifier makes a mistake (predicting 1 when the class is 0 or 0 when the class is 1) is $p_0$. Now, you make an ensamble of the 5 classifiers in such a way that for a particular sample each classifier is applied and each one makes a vote. A sample will be classified as 1 if 3 or more classifiers predict 1, otherwise it will classifed as 0.

Given $p_0$ calculate the probability that the ensamble makes a mistake. 

````python
def p3(p0):
    '''
    p0: probability that an individual classifier makes a mistake
    
    returns:
    Probability that the ensamble makes a mistake
    '''
    ### your code here
    #El problema sigue una distribución binomial, la muestra (sample) marcara error si suceden 3 o mas errores, es decir,
    #3 o mas equivocaciones, cada una con una probabilidad p0 que suceda, siguiendo la distribución binomial, se tendria que
    #se estan haciendo 5 ensayos de Bernoulli (n=5), con probabilidad p0 y maracara error con 3,4 o 5 veces (k=[3 4 5])
    n=5
    k=[3,4,5]
    p3=nCr(n,k[0])*p0**k[0]*(1-p0)**(n-k[0])
    p4=nCr(n,k[1])*p0**k[1]*(1-p0)**(n-k[1])
    p5=nCr(n,k[2])*p0**k[2]*(1-p0)**(n-k[2])
    res = p3+p4+p5
    return res
    
````

````python
#TEST_CELL
#Se define una función para calcular la combinatoria de 2 números
import math
def nCr(n,r):
    f = math.factorial
    return f(n) / f(r) / f(n-r)
````
